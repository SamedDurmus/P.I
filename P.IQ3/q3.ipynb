```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel ('daily.xlsx')
```

```python
df
```

```python
data=df.Work.mean(),df.Family.mean(),df.Homeworks.mean(),df.Individual.mean(),df.Socializing.mean(),df.SpareTime.mean(),df.Sleep.mean()
labels= ["Work", "Family","Homeworks","Individiual","Socializing","Spare Time", "Sleep"]
```

```python
data
```

```python
 
colors = ["orange", "cyan", "brown", "grey", "indigo", "beige","blue"] 
explode = [0,0,.1,.2,.4,.5,.6]

fig1, ax1 = plt.subplots(figsize = (8, 8))

patches, texts, autotext = ax1.pie(data, colors = colors, startangle=90, autopct='%0.2f%%', explode = explode, textprops={'fontsize': 20})

ax1.axis("equal")
plt.tight_layout() 
plt.legend(labels,loc= 2,borderpad=0.05, fontsize = 14, bbox_to_anchor=(1,1))
plt.title('\n Distribution of Daily activities \n(for Charles,Henry and Susan by the averages)', fontsize = 20)
plt.savefig("piechart_daily.png", bbox_inches="tight")

plt.show()

```
